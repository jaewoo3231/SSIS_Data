```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
from matplotlib import cm
import matplotlib.font_manager as fm
import matplotlib
%matplotlib inline
plt.style.use('ggplot')

font_location = 'C:/Windows/Fonts/malgunsl.ttf'
font_name = fm.FontProperties(fname = font_location).get_name()
matplotlib.rc('font' , family=font_name)

from sklearn.feature_selection import RFE
from sklearn.linear_model import LogisticRegression

import warnings
warnings.filterwarnings(action='ignore')
```

```python
Child_Abuser = pd.read_csv("./Child_Abuser_dum.csv",encoding = 'CP949')
Child_Abuser
```

```python
for i in range(len(Child_Abuser.columns)):
    print(i, Child_Abuser.columns[i])
```

```python
for i in Child_Abuser:
    print(i,Child_Abuser[i].unique(),'\n')
```

```python
for i in Child_Abuser['조대_재신고여부_1'].index:
    if (Child_Abuser['조대_재신고여부_1'][i] == 'Y'):
        Child_Abuser['조대_재신고여부_1'][i] = 1
    elif (Child_Abuser['조대_재신고여부_1'][i] == 'N'):
        Child_Abuser['조대_재신고여부_1'][i] = 0
```

```python
X = Child_Abuser.drop(['New_재학대혐의여부','신고_접수연도'], axis=1)
y = Child_Abuser['New_재학대혐의여부']
```

```python
# 15-20분쯤 걸림
model = LogisticRegression()
rfe = RFE(model, 20)
fit = rfe.fit(X, y)

print("Num Features: %d" % fit.n_features_)
print("Selected Features: %s" % fit.support_)
print("Feature Ranking: %s" % fit.ranking_)
```

```python
for i in range(len(X.columns)):
    print(i, X.columns[i], fit.support_[i], fit.ranking_[i])
```

```python
for i in range(len(X.columns)):
    if fit.ranking_[i] == 1:
        print(X.columns[i])
```
