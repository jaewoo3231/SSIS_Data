```python
for i in df3.index:
    try:
        t1 = str(int(df3['결과_조치결과일자'][i])) # 조치일자 
        t1 = t1[:4]+'-'+t1[4:6]+'-'+t1[6:8]
        t1 = datetime.strptime(t1,'%Y-%m-%d')
        t2 = datetime.strptime(df3['신고_접수일시'][i], '%Y-%m-%d %H:%M:%S') # 접수일자
        result = t1-t2
        if(int(str(result).split('days')[0].split('day')[0])==-1): # 당일
            df3['New_조치걸린일수'][i] = 0
        elif(int(str(result).split('days')[0]) > 0): # 음수제외 일수 삽입
            df3['New_조치걸린일수'][i] = int(str(result).split('days')[0])
        else: # 음수일때 nan 삽입
            df3['New_조치걸린일수'][i]=np.nan 
    except ValueError:
        if(result<=timedelta(days=1)): # 다음날 조치
            df3['New_조치걸린일수'][i]=1
            continue
        df3['New_조치걸린일수'][i]=np.nan # 그외 에러일때 nan 
        
```
