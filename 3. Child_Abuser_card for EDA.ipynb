```python
# int(Abuser['행위_생년월일'][0][:2]) # 생년월일
# Child['Unnamed: 4'][5543]
len(Abuser['행위_생년월일'][10000])==6
```

```python
# 접수일시에서의 연도에서 생년월일의 생년을 빼서 나온수치를 행위자의 나이로 사용할 예정
for i in range(len(Abuser)):
    try:
        age = int(str(Abuser['신고_접수일시'][i])[2:4]) - int(Abuser['행위_생년월일'][i][:2])
        if (age < 0):
            age = 100 + int(str(Abuser['신고_접수일시'][i])[2:4]) -  int(Abuser['행위_생년월일'][i][:2])
#         if ((0<= age <=17) != True):
#             age = np.nan
        if (len(Abuser['행위_생년월일'][i]) ==6):
            Abuser['행위_생년월일'][i] = age
        else:
            Abuser['행위_생년월일'][i] = np.nan
    except TypeError:
        Abuser['행위_생년월일'][i] = np.nan
    except ValueError:
        Abuser['행위_생년월일'][i] = np.nan
        
```

```python
print(Abuser['행위_생년월일'].value_counts())
print("null 값: ",Abuser['행위_생년월일'].isnull().sum() )
```

## 1-2. 나이 연령 범주형으로 바꾸기
### 19세이하 , 20대(20~29세),30대(30~39세),40대(40~49세), 50대(50~59세),60대(60~69세),70세 이상, 파악불가
### 2018 아동학대 주요통계 참고

```python
Abuser['행위_생년월일']
```

```python
for i in range(len(Abuser['행위_생년월일'])):
    try:
        if (0<= Abuser['행위_생년월일'][i] <= 19 ):
            Abuser['행위_생년월일'][i] = '19세이하'
        elif (20<= Abuser['행위_생년월일'][i] <= 29):
            Abuser['행위_생년월일'][i] = '20대'
        elif (30<= Abuser['행위_생년월일'][i] <= 39):
            Abuser['행위_생년월일'][i] = '30대'
        elif (40<= Abuser['행위_생년월일'][i] <= 49):
            Abuser['행위_생년월일'][i] = '40대'
        elif (50<= Abuser['행위_생년월일'][i] <= 59):
            Abuser['행위_생년월일'][i] = '50대'
        elif (60<= Abuser['행위_생년월일'][i] <= 69):
            Abuser['행위_생년월일'][i] = '60대'
        elif (70<= Abuser['행위_생년월일'][i]):
            Abuser['행위_생년월일'][i] = '70세이상'
    except ValueError:
        continue
```
