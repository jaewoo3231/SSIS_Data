```python
#학대 유형 분류
abuse_cat = []

for i in range(0, len(Child_Abuser)):
    abuse_cat.insert(i, None)

abuse_cat

for i in range(0, len(Child_Abuser)):
    if Child_Abuser['판단_유형_신체'][i] == 'Y':
        abuse_cat[i] = '신체'

    if Child_Abuser['판단_유형_정서'][i] == 'Y':
        if abuse_cat[i] == None:
            abuse_cat[i] = '정서'
        else:
            abuse_cat[i] = abuse_cat[i] + ',정서'

    if Child_Abuser['판단_유형_성'][i] == 'Y':
        if abuse_cat[i] == None:
             abuse_cat[i] = '성'
        else:
             abuse_cat[i] = abuse_cat[i] + ',성'

    if Child_Abuser['판단_유형_방임'][i] == 'Y':
        if abuse_cat[i] == None:
             abuse_cat[i] = '방임'
        else:
             abuse_cat[i] = abuse_cat[i] + ',방임'

    if Child_Abuser['판단_유형_유기'][i] == 'Y':
        if abuse_cat[i] == None:
             abuse_cat[i] = '유기'
        else:
             abuse_cat[i] = abuse_cat[i] + ',유기'

# abuse_cat
```
